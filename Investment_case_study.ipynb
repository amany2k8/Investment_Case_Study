```python
import warnings
import itertools
import numpy as np
```

```python
import pandas as pd
files=pd.read_csv('grocers_t2_C01.csv')  
```

```python
files
```

```python
files['day'] = pd.to_datetime(files['day']) # Convert the day column into pandas datetime type
```

```python
files['quarter'] = files['day'].dt.to_period('Q-DEC')  # Convert the transformed column quarter with year(Q-DEC means taking december as year end)
```

```python
files
```

```python
cum_columns = ['devices']

result = files.merge(
    files.groupby('id')[cum_columns].diff(),
    left_index=True, right_index=True, suffixes=['', '_uncum']
).fillna({'{}_uncum'.format(cum_column): files[cum_column] for cum_column in cum_columns})

# Thought the devices were getting added with every day of observation, so created an uncummulative column
```

```python
result
```

```python
df=(result.groupby(['quarter','id','Store Name','Address','Retailer'])
    .mean()
    .reset_index()
)
```

```python
df['norm_traffic']=df['devices_store'] /df['devices']  # Normalized traffic by first grouping by 'quarter','id','Store Name','Address','Retailer'
```

```python
df
```

2.	For each grocer, which store has the highest traffic (and how many devices visited) and which store has the lowest traffic each quarter?

```python
grocer=df.Retailer.unique() # found out the unique grocer names so that could loop around for finding store for that grocer with highest and lowest traffic
```

```python
grocer
```

```python
final = pd.DataFrame(columns=['quarter','id','Store Name','Address','Retailer','devices_store','devices','traffic','devices_uncum','norm_traffic']) # creating empty set with column name to append 
for Retailer in grocer:
    bp_df=df.loc[df['Retailer']==Retailer]
    bp_df=bp_df.loc[bp_df["norm_traffic"]==bp_df["norm_traffic"].max()] # Subsetting using loc(label based) with max norm traffic
    final = pd.concat([final, bp_df], ignore_index=True)
final
```

```python
final.to_excel("Q2.xlsx")
```

```python
final33 = pd.DataFrame(columns=['quarter','id','Store Name','Address','Retailer','devices_store','devices','traffic','devices_uncum','norm_traffic'])
for Retailer in grocer:
    bp_df33=df.loc[df['Retailer']==Retailer]
    bp_df33=bp_df33.loc[bp_df33["norm_traffic"]==bp_df33["norm_traffic"].min()]
    final33 = pd.concat([final33, bp_df33], ignore_index=True)
final33
```

```python
final33.to_excel("Q2_Low.xlsx")
```

3.	For each grocer, which store had the highest year-over-year growth (i.e. 1Q2017 vs. 1Q2016 growth) in device traffic?

```python
dpq2=df.drop(['Address','traffic','devices_uncum'], axis=1)
```

```python
dl22=dpq2.groupby(['Retailer','quarter','Store Name'], as_index=False).mean()
dl22['pct_%'] = dl22['norm_traffic'].pct_change(4) # finding out % change after every 4 quarters/ 12 months.
dl22
```

```python
dl22.to_excel("query_12.xlsx")
```

```python
final70 = pd.DataFrame(columns=['Retailer','quarter','id','Store Name','devices_store','devices','norm_traffic','pct_%','rpct_%'])
for Retailer in grocer:
    bp_df18=dl22.loc[dl22['Retailer']==Retailer]
    id_r=bp_df18.id.unique()
    final7=pd.DataFrame(columns=['Retailer','quarter','id','Store Name','devices_store','devices','norm_traffic','pct_%','rpct_%'])
    for ids in id_r:
        b_df18=bp_df18.loc[bp_df18['id']==ids]
        b_df18['rpct_%'] = b_df18['norm_traffic'].pct_change(4)
        b_df18=b_df18.loc[b_df18["rpct_%"]==b_df18["rpct_%"].max()]
        final7 = pd.concat([final7, b_df18], ignore_index=True)
    final7=final7.loc[final7["rpct_%"]==final7["rpct_%"].max()]
    final70=pd.concat([final70, final7], ignore_index=True)
final70
```

```python
final70.to_excel("Q3.xlsx")
```
